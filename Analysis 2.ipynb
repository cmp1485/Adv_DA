```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats as st
import seaborn as sns
```

```python
df= pd.read_csv("Ger_autos_cleaned.csv")
```

#### 1) Can you tell me No of Vehicles by Brand Available on ebay for sale with the help of visualization

```python
brand_cnt = df['brand'].value_counts()
brand_cnt
```

```python
brand_cnt.plot(kind='bar')
```

#### 2) What is the Average price for vehicles based on the type of vehicle as well as on the type of gearbox.Explain me with both numerical and visualization analysis

```python
ct = pd.crosstab(index=df['vehicleType'], columns = df['gearbox'], values=df['price'], aggfunc='mean')
ct
```

```python
ct.plot(kind='bar')
```

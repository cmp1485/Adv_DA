```python
df = df.drop(columns=['notRepairedDamage', 'nrOfPictures'])
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df = df.drop(df[(df['yearOfRegistration'] < 1900) | (df['yearOfRegistration'] >2023)].index)
```

```python
df['powerPS'] = np.where((df['powerPS'] < 1) | (df['powerPS'] >2400), df['powerPS'].median(), df['powerPS'])
```

```python
df['monthOfRegistration'] = np.where((df['monthOfRegistration'] == 0), df['monthOfRegistration'].mode()[0], df['monthOfRegistration'])
```

```python
df = df.drop(df[(df['price']<500) | (df['price']>4000000)].index)
```

```python
df.drop_duplicates(inplace=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ger_autos_cleaned.csv")
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

#### 1) The memory usage of the data is around 6.1 mb.How can we reduce the memory usage of the data set?

*downcasting- find the range of values in a column and reduce int64 and float64 columns to sub-datatypes that occupy lesser space
*drop columns that have zero variance or are unnecessary for our analysis

#### 2) What is the Average price of vehicle by fuel type and gearbox type.Give a plot

```python
fg = pd.crosstab(columns=df['gearbox'], index = df['fuelType'], values=df['price'], aggfunc='mean')
```

```python
fg
```

```python
fg.plot(kind='bar')
```

#### 3) What is the Average power of a vehicle by vehicle type and gearbox type.Give a plot

```python
vg = pd.crosstab(columns=df['gearbox'], index = df['vehicleType'], values=df['powerPS'], aggfunc='mean')
```

```python
vg
```

```python
vg.plot(kind='bar')
```

#### 4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this

```python
bv = pd.crosstab(index=df['brand'], columns = df['vehicleType'], values=df['price'], aggfunc='mean')
```

```python
bv
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(bv, annot=True, linewidths=.5, ax=ax)
```
